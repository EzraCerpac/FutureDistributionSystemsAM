## Section 1:/ Introduction 

The operation of electrical distrubution grids is changing fast. The infeed of renewables and the loading by battery charging is causing unforeseen challenges. The finite element modeling of electrical power transformer is indispensable to chart the amplitude, frequency content and location of electromagnetic losses.

For more information, see [master thesis text of Max van Dijk](https://repository.tudelft.nl/islandora/object/uuid%3A15b25b42-e04b-4ff2-a187-773bc170f061?collection=education).   

We consider the same three-phase power transformer considered earlier in the master thesis of Max van Dijk. (Need to include more specs here). 

We wish to develop a model consisting of the following four components: 
1. a model in which the <b>magnetic field</b> is solved coupled with the <b>electrical circuit</b> for the currents in the windings. It is assumed that three-dimensional end effects can be neglected die to the lamination of the ferromagnetic core. A two-dimensional perpendicular current model for the magnetic field will therefore be solved. The magnetic field will be solved in terms of the perpendicular component of the vector potential as before in the course; 
2. a model for the estimation of the <b>iron losses</b> in the ferromagnetic core and the <b>copper losses</b> in the windings; 
3. a model for the <b>heat source</b> due to the transformer losses; 
4. a model for the <b>temperature</b> of the substation in which the transformer is housed; 

It is expected that the introduction of new sources (e.g. solar panels) and loads (e.g. electrical vehicle charging) in the distribution grid which causes higher frequency harmonics. Of particular interest is the degradation of the thermal insulating properties of the transformer and the reduction of the life-time of the transformer over time due to higher harmonics.

De volgende stap hangt heel erg af van het doel.

Als de wens is om met simpele modellen de verliezen (en later evt. temperatuurverdeling) in de trafo te modelleren, dan denk ik dat de volgende stap moet zijn om netjes alle materiaalparameters te bepalen.

De BH curve in mijn model is vrij eenvoudig (gebaseerd op een eenvoudige formule), terwijl deze in werkelijkheid misschien anders is.

Verder zijn de Steinmetz parameters uit Max’ thesis naar mijn idee helemaal niet juist. Als we deze goed kunnen bepalen moet het mogelijk zijn een stuk dichter bij de gemeten verliezen te komen, zonder dat er complexe 3D modellen gemaakt hoeven te worden.
 
Misschien het meest belangrijke, als we Gridap verder willen gebruiken, is de ontwikkeling van een betere interface voor de field-circuit coupling. Op dit moment is het door mijn implementatie namelijk niet mogelijk om gebruik te maken van sommige Gridap-onderdelen (zoals FEOperators), en daarmee wordt het modelleren van multi-field en transiente problemen een stuk lastiger. Dit zou bijvoorbeeld een extensie voor Gridap kunnen worden. De meeste code wordt op dit moment namelijk “verspild” aan de herhaaldelijke definitie van allerlei dingen in Gridap (integralen over de wikkelingen, voor de field-circuit coupling). Dit zou de code voor zowel TA1 als TA2 een stuk leesbaarder maken en fijner om mee te werken.
